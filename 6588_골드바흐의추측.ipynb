```
import math

def isprime(x):
    for i in range(2, int(math.sqrt(x))+1):
      if x % i == 0:
        return False
    return True
  
while(1):
  n = int(input())
  if n == 0:
    break

  a = []
  b = []
  #c = []
  if (n//2)%2 == 1:
    for i in reversed(range(0, n//2 - 1, 2)): #3까지
      if isprime(n//2 - i) and isprime(n//2 + i) == 1:
        a.append(n//2 - i)
        b.append(n//2 + i)
        break
  else: 
    for i in reversed(range(1, n//2 - 1, 2)): #3까지
      if isprime(n//2 - i) and isprime(n//2 + i) == 1:
        a.append(n//2 - i)
        b.append(n//2 + i)
        break
  if len(a)==0:
    print("Goldbach's conjecture is wrong.")
  else:
    print(n,"=",a[0],"+",b[0])
```
