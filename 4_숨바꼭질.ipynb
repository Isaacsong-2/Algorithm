```
n, k = map(int, input().split())
```

```
def wfs(n, k):
  queue = [n]
  sol = [100001]*100001
  sol[n] = 0
  while queue:
    a = queue.pop(0)
    #del queue[0]
    
    if a == k:
      print(sol[a])
      break
    
    if a < 100000 and sol[a+1] > sol[a] + 1:
      sol[a+1] = sol[a] + 1
      queue.append(a+1)
    
    if a >0 and sol[a-1] > sol[a] + 1:
      sol[a-1] = sol[a] + 1
      queue.append(a-1)

    if a <= 50000 and sol[2*a] > sol[a] + 1:
      sol[2*a] = sol[a] + 1
      queue.append(2*a)
    #print(queue)
wfs(n, k)
#print(n)
#print(a)
#print(k)



```
