```
import math
from itertools import combinations
N = int(input())
S = ([list(map(int, input().split())) for _ in range (N)])
```

```
m = []
c = list(combinations(range(N),N//2)) #이런방법이?
print(c)
for i in range (int(len(c)/2)):
  a = list(set(c[i]))
  b = list(set(c[int(len(c))-i-1]))
  t1, t2 = 0, 0
  print(a,b)
  for j in range (N//2):
    for k in range (j, N//2):
      t1 += S[a[j]][a[k]] + S[a[k]][a[j]]
      t2 += S[b[j]][b[k]] + S[b[k]][b[j]]
  print(t1,t2)
  m.append(abs(t1 - t2))
print(min(m))
#
```
