```
n = int(input())
st = []
rs = []
count = 1
temp = True
for i in range(n):
    num = int(input())
    while count <= num:
        st.append(count)
        rs.append('+')
        count += 1
    if st[-1] == num:
        st.pop()
        rs.append('-')
    else:
        temp = False
if temp == False:
    print('NO')
else:
    for i in rs:
        print(i)
```
