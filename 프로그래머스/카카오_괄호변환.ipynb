```python
def solution(p):
  answer = ''
  if len(p) == 0: return ''
  lcnt, rcnt = 0, 0
  for i in range(len(p)):
    if p[i] == '(':
      lcnt += 1
    if p[i] == ')':
      rcnt += 1
    if lcnt == rcnt:
      u = p[:i+1]
      v = p[i+1:]
      break
  lcnt, rcnt = 0,0
  for i in range(len(u)):
    if p[i] == '(':
      lcnt += 1
    if p[i] == ')':
      rcnt += 1
    if rcnt > lcnt: #u가 올바르지 않은 문자열
      answer = answer + '('
      answer = answer + solution(v)
      answer = answer + ')'
      for j in range(1, len(u)-1):
        if u[j] == '(':
          answer = answer + ')'
        else:
          answer = answer + '('
      return answer
  answer = answer + u + solution(v)
  return answer
```

```python
print(solution(p))
```
