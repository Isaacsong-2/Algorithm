```python
n,k = map(int,input().split())
t = [[0,0]]
for i in range(n):
  t.append(list(map(int, input().split())))
```

```python
print(t)
```

```python
dp = [[0] * (k+1) for _ in range(n+1)]
sum = 0
for i in range(1, n+1):
  for j in range(1, k+1):
    w = t[i][0]
    v = t[i][1]
    if j < w:
      dp[i][j] = dp[i-1][j]
    else:
      dp[i][j] = max(dp[i-1][j], dp[i-1][j-w]+v)
print(dp[n][k])
  
  
```
