```
n = int(input())
h = [list(map(int, input().split())) for _ in range(n)]
```

```
import copy
t = copy.deepcopy(h)
for i in range(1, n):
  t[i][0] = t[i][0] + min(t[i-1][1], t[i-1][2])
  t[i][1] = t[i][1] + min(t[i-1][0], t[i-1][2])
  t[i][2] = t[i][2] + min(t[i-1][0], t[i-1][1])
print(min(t[n-1]))
```
