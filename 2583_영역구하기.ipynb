```
m, n, k = [int(_) for _ in input().split()]
n_maps = [[0]*n for _ in range(m)]
for i in range(k):
  x1, y1, x2, y2 = [int(_) for _ in input().split()]
  for p in range(y1, y2):
    for q in range(x1, x2):
      n_maps[p][q] = 1
```

```
print(n_maps)
```

```

dx = [1, -1, 0, 0]
dy = [0, 0, 1, -1]

import copy
from collections import deque
maps = copy.deepcopy(n_maps)
queue = deque()

def bfs(s):
  queue = [s]
  cnt = 1
  maps[s[0]][s[1]] = 2
  while queue:
    a, b = queue.pop()
    for i in range(4):
      x = a + dx[i]
      y = b + dy[i]
      if 0 <= x < m and 0 <= y < n and maps[x][y] == 0:
        maps[x][y] = 2
        cnt += 1
        queue.append((x,y))
  return cnt
  
sol = []
for i in range(m):
  for j in range(n):
    if maps[i][j] == 0:
      sol.append(bfs((i,j)))
sol.sort()
print(len(sol))
for k in sol:
  print(k, end = " ")
```
