```
n = int(input())
dx = [1, -1, 0, 0]
dy = [0, 0, -1, 1]
m = []
cnt = []
for i in range(n):
  m.append(list(input()))
```

```
def bfs(p, q):
  queue = [[p, q]]
  c = 1
  m[p][q] = "0"
  while queue:
    a, b = queue[0][0], queue[0][1]
    del queue[0]
    for k in range(4):
      x = a + dx[k]
      y = b + dy[k]
      if 0 <= x < n and 0 <= y < n and m[x][y] == "1":
        m[x][y] = "0"
        queue.append([x, y])
        c += 1
  cnt.append(c)
for i in range(n):
    for j in range(n):
        if m[i][j] == "1":
            bfs(i, j)
cnt.sort()
print(len(cnt))
for i in cnt:
  print(i)
```
