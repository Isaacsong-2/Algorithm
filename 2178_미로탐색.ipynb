```
n, m = map(int, input().split())
s = []
for i in range(n):
  s.append(list(input()))
```

```
dx = [1, -1, 0, 0]
dy = [0, 0, -1, 1]
def wfs(n, m):
  queue = [(0, 0)]
  c = [[0]*m for _ in range(n)]
  while queue:
    #print(queue)
    a, b = queue.pop(0)
    if a == n-1 and b == m-1:
      print(c[a][b]+1)
      break
    for k in range(4):
      x = a + dx[k]
      y = b + dy[k]
      if 0 <= x < n and 0 <= y < m and s[x][y] == '1' and c[x][y] == 0:
        #s[x][y] = '0'
        queue.append((x, y))
        c[x][y] = c[a][b] + 1

wfs(n,m)
#print(s)
  
```
