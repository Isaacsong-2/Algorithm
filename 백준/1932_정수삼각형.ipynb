```python
n = int(input())
t = [list(map(int, input().split())) for _ in range(n)]
```

```python
dp = list([0]*n for _ in range(n))
dp[0][0] = t[0][0]
for i in range(n-1):
    for j in range(i+2):
        dp[i+1][j] = t[i+1][j] + max(dp[i][j],dp[i][j-1])
#print(dp)
print(max(dp[n-1]))
```

```python
if n==1:
  print(t[0][0])
else:
  k = 2
  for i in range(1,n):
    for j in range(k):
      if j == 0:
        t[i][j] = t[i][j] + t[i-1][j]
      elif i == j:
        t[i][j] = t[i][j] + t[i-1][j-1]
      else:
        t[i][j] = max(t[i - 1][j - 1], t[i - 1][j]) + t[i][j]
    k +=1
print(max(t[n-1]))
```
