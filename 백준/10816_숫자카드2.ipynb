```
'''
N = int(sys.stdin.readline())
a = list(map(int, sys.stdin.readline().split()))
M = int(sys.stdin.readline())
b = list(map(int, sys.stdin.readline().split()))
'''
N = int(input())
a = list(map(int, input().split(' ')))
M = int(input())
b = list(map(int, input().split(' ')))
count = [0]*M
i = 0
j = 0
while(i != M):
  if b[i] in a:
    count[i] +=1
    del a[a.index(b[i])]
  else:
    i +=1
for i in count:
    print(i, end = ' ')
```
