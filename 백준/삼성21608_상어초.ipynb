```python
n = int(input())
t = [list(input().split()) for _ in range(n*n)]
```

```python
print(t)
```

```python
from collections import deque
visit = [[0]*n for _ in range(n)]
dx = [1, -1, 0, 0]
dy = [0, 0, 1, -1]


for st in range(n*n):
  near = [[0]*n for _ in range(n)]
  rc = [[0]*n for _ in range(n)]
  for i in range(n):
    for j in range(n):
      if visit[i][j] == 0:
        for k in range(4):
          x = i + dx[k]
          y = j + dy[k]
          if (0 <= x < n and 0 <= y < n and (visit[x][y] == int(t[st][1]) or visit[x][y] == int(t[st][2]) or visit[x][y] == int(t[st][3]) or visit[x][y] == int(t[st][4]))):
            near[i][j] += 1

  #비어있는 칸 중에서 좋아하는 학생이 인접한 칸에 가장 많을때 max
  max = -1
  for i in range(n):
    for j in range(n):
      if visit[i][j] == 0:
        if near[i][j] > max:
          max = near[i][j]
          queue = deque()
          queue.append((i, j))
        elif near[i][j] == max:
          queue.append((i, j))

  #1을 만족하는 칸이 queue에 들어가고 비어있는 칸 계산
  #print(queue)
  if (len(queue) == 1):
    a, b = queue.popleft()
    visit[a][b] = int(t[st][0])
  else:
    while queue:
      a, b = queue.popleft()
      #if (queue == 0):
      #   visit[a][b] = int(t[st][0])
      #   break
      for k in range(4):
        x = a + dx[k]
        y = b + dy[k]
        if (0 <= x < n and 0 <= y < n and visit[x][y] == 0):
          rc[a][b] += 1
    max = -1
    for i in range(n):
      for j in range(n):
        if rc[i][j] > max:
          max = rc[i][j]
          queue = deque()
          queue.append((i, j))
        elif rc[i][j] == max:
         queue.append((i, j))
    #if max == 0:
    #  queue = deque()
    #print(queue)

  #2를 만족하는 칸 여러개
    min_row = 21
    queue2 = deque()
    while queue:
      a, b = queue.popleft()
      #if queue==0:
      #  break
      if b < min_row:
        min_row = b
        queue2 = deque()
        queue2.append((a,b))
      elif b == min_row:
        queue2.append((a,b))
    min_col = 21
    while queue2:
      a, b = queue2.popleft()
      #if queue2 == 0:
      #  break
      if a < min_col:
        min_col = a
    visit[min_col][min_row] = int(t[st][0])
print(visit)
sol = 0
for i in range(n):
  for j in range(n):
    count = 0
    for k in range(4):
      x = i + dx[k]
      y = j + dy[k]
      if (0 <= x < n and 0 <= y < n and (visit[x][y] == int(t[n*i+j][1]) or visit[x][y] == int(t[n*i+j][2]) or visit[x][y] == int(t[n*i+j][3]) or visit[x][y] == int(t[n*i+j][4]))):
        count += 1
    if count == 1: sol += 1
    if count == 2: sol += 10
    if count == 3: sol += 100
    if count == 4: sol += 1000
    print(sol)
  



    
  


  



  
```

```python
import sys
from collections import defaultdict
input = sys.stdin.readline
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
N = int(input())
arr = [[0]*N for _ in range(N)]
likedict = defaultdict(list)
result = 0
 
for _ in range(N*N):
    _input = list(map(int, input().split()))
    likedict[_input[0]] = _input[1:]
    
    max_x = 0
    max_y = 0
    max_like = -1
    max_empty = -1
    for i in range(N):
        for j in range(N):
            if arr[i][j] == 0:
                likecnt = 0
                emptycnt = 0
                for k in range(4):
                    nx = i + dx[k]
                    ny = j + dy[k]
                    if 0 <= nx < N and 0 <= ny < N:
                        if arr[nx][ny] in _input:
                            likecnt += 1
                        if arr[nx][ny] == 0:
                            emptycnt += 1
                            
                if max_like < likecnt or (max_like == likecnt and max_empty < emptycnt):
                    max_x = i
                    max_y = j
                    max_like = likecnt
                    max_empty = emptycnt
                    
    arr[max_x][max_y] = _input[0]
    
for i in range(N):
    for j in range(N):
        cnt = 0
        like = likedict[arr[i][j]]
        for k in range(4):
            nx = i + dx[k]
            ny = j + dy[k]
            if 0 <= nx < N and 0 <= ny < N:
                if arr[nx][ny] in like:
                    cnt += 1
        if cnt != 0:
            result += 10 ** (cnt-1)
            
print(result)
```
