```python
import math
n = int(input())
a = [0]*n
for i in range (n):
  a[i] = int(input())
```

```python
a.sort()
b = []
for i in range(n-1):
  b.append(a[i+1]-a[i])
#print(b)
temp = math.gcd(b[0], b[1])
for i in range(1, len(b)-1):
  temp = math.gcd(temp,b[i+1])
#print(temp)
#print(math.gcd(b[0],b[1]))
sol = 0
for i in range(len(b)):
  sol += b[i]/temp - 1
print(int(sol))




```
