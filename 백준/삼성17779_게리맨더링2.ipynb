```python
n = int(input())
a = [[]]
for i in range(n): a.append([0]+ list(map(int,input().split())))
```

```python
def div(x,y,d1,d2):
  fifth = [[0] * (n+1) for _ in range (n+1)]
  fifth[x][y] = 5
  for i in range(1, d1 + 1):
    fifth[x + i][y - i] = 5
  for i in range(1, d2 + 1):
    fifth[x + i][y + i] = 5
  for i in range(1, d2 + 1):
    fifth[x + d1 + i][y - d1 + i] = 5
  for i in range(1, d1 + 1):
    fifth[x + d2 + i][y + d2 - i] = 5
  
  temp = [0, 0, 0, 0, 0]
  for r in range(1, x + d1):
    for c in range(1, y + 1):
      if fifth[r][c] == 5:
        break
      else:
        temp[0] += a[r][c]
  for r in range(1, x + d2 + 1):
    for c in range(n, y, -1):
      if fifth[r][c] == 5:
        break
      else:
        temp[1] += a[r][c]
  for r in range(x + d1, n + 1):
    for c in range(1, y - d1 + d2):
      if fifth[r][c] == 5:
        break
      else:
        temp[2] += a[r][c]
  for r in range(x + d2 + 1, n + 1):
    for c in range(n, y - d1 + d2 - 1, -1):
      if fifth[r][c] == 5:
        break
      else:
        temp[3] += a[r][c]
  temp[4] = total - sum(temp)
  return (max(temp) - min(temp))

total = 0
for i in range(n+1):
  total += sum(a[i])
sol = 10000000
for x in range(n+1):
  for y in range(n+1):
    for d1 in range(n+1):
      for d2 in range(n+1):
        if 1 <= x < x + d1 + d2 <= n and 1 <= y - d1 < y < y + d2 <= n:
          sol = min(sol, div(x,y,d1,d2))
```
