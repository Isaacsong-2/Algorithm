```python
import sys
import heapq
input = sys.stdin.readline
n = int(input())
arr_l = []
arr_r = []
answer = []
for i in range(n):
  num = int(input())
  if len(arr_l) == len(arr_r): #길이가 같으면 왼쪽큐
    heapq.heappush(arr_l, [-num,num])
  else:
    heapq.heappush(arr_r, num) #아니면 오른쪽큐
  if arr_r and arr_l[0][1] > arr_r[0]: # 왼쪽큐 최대값이 오른쪽큐의 최소값보다 크면
    m = heapq.heappop(arr_r)
    M = heapq.heappop(arr_l)
    heapq.heappush(arr_l,[-m,m]) #서로 바꿔줌
    heapq.heappush(arr_r,M[1])
  answer.append(arr_l[0][1])
for i in answer:
  print(i)

# for i in range(n):
#   if i == 0:
#     heapq.heappush(arr_l,int(input()))
#     print(arr_l[0])
#     continue
#   if i == 1:
#     num = int(input())
#     if arr_l[0] > num:
#       heapq.heappush(arr_r, heapq.heappop(arr_r))
#       heapq.heappush(arr_l, num)
#     else:
#        heapq.heappush(arr_r, num)
#     print(arr_l[0])
#     continue
#   num = int(input())
#   if len(arr_l) == len(arr_r):
#     if arr_r[0] < num:
#       heapq.heappush(arr_r, num)
#       heapq.heappush(arr_l, heapq.heappop(arr_r))
#     else:
#       heapq.heappush(arr_l, num)
#   else:
#     if arr_r[0] < num:
#       heapq.heappush(arr_r, num)
#     else:
#       heapq.heappush(arr_l, num)
  # print("-------",i,"-------")
  # print(arr_l, arr_r)
  # print(max(arr_l))
```
