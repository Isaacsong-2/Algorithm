```python
n,k,h,w,d = map(int,input().split())
li = []
h_min= 100000
h_max = 0
w_min = 100000
w_max = 0
for i in range(n):
    hlo, hhi, wlo, whi = map(int,input().split())
    h_min = min(h_min,hlo)
    h_max = max(h_max, hhi)
    w_min = min(w_min,wlo)
    w_max = max(w_max,whi)
    li.append([hlo,hhi,wlo,whi])
# li = [list(map(int,input().split())) for _ in range(n)]
```

```python
h_min_ran = -min(d, h-h_min)
h_max_ran = min(d, h_max-h)
w_min_ran = -min(d,w-w_min)
w_max_ran = min(d,w_max-w)
sol = 0
for i in range(h_min_ran,h_max_ran+1):
    for j in range(w_min_ran,w_max_ran+1):
        cnt = 0
        for num in range(n):
            if li[num][0] <=h+i<= li[num][1] and li[num][2] <=w+j<= li[num][3]:
                cnt += 1
            if cnt==3:
                sol += 1
                break
print(sol)
```
