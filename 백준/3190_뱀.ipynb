```python
n = int(input())
k = int(input())
li = [[0] * n for _ in range(n)]
for i in range(k):
    r, c = map(int,input().split())
    li[r-1][c-1] = 1
l = int(input())
sec = [0]
turn = []
for i in range(l):
    s, d = input().split()
    s = int(s)
    sec.append(s)
    turn.append(d)
```

```python
import copy
apple = copy.deepcopy(li)
dir = ['r','u','d','l']
cur_dir = 'r'
snake = [[0] * n for _ in range(n)]
snake[0][0] = 1
sol = 0
def check(cur_dir):
    r = 0
    c = 0
    tail = [[0, 0]]
    global sol
    for i in range(l):
        for j in range(sec[i+1]-sec[i]):
            sol += 1
            if cur_dir == dir[0]:
                nc = c+1
                nr = r
            elif cur_dir == dir[1]:
                nr = r-1
                nc = c
            elif cur_dir == dir[2]:
                nr = r+1
                nc = c
            elif cur_dir == dir[3]:
                nc = c-1
                nr = r
            if 0<=nr<n and 0<=nc<n and snake[nr][nc] == 0:
                r = nr
                c = nc
                snake[r][c] = 1
                tail.append([r,c])
                if apple[r][c] == 1:
                    apple[r][c] = 0
                else:
                    tr, tc = tail.pop(0)
                    snake[tr][tc] = 0
            else:
                return sol
            # if(i>0):
            #     print(i)
            #     print(cur_dir,j,sec[i+1]-sec[i])
            #     print(snake)
            #     print(tail)
        if cur_dir == 'r':
            if turn[i] == 'D':
                cur_dir = 'd'
            else: cur_dir = 'u'
        elif cur_dir == 'l':
            if turn[i] == 'D':
                cur_dir = 'u'
            else: cur_dir = 'd'
        elif cur_dir == 'u':
            if turn[i] == 'D':
                cur_dir = 'r'
            else: cur_dir = 'l'
        elif cur_dir == 'd':
            if turn[i] == 'D':
                cur_dir = 'l'
            else: cur_dir = 'r'
    while(1):
        sol += 1
        if cur_dir == dir[0]:
            nc = c+1
            nr = r
        elif cur_dir == dir[1]:
            nr = r-1
            nc = c
        elif cur_dir == dir[2]:
            nr = r+1
            nc = c
        elif cur_dir == dir[3]:
            nc = c-1
            nr = r
        if 0<=nr<n and 0<=nc<n and snake[nr][nc] == 0:
            r = nr
            c = nc
            snake[r][c] = 1
            tail.append([r,c])
            if apple[r][c] == 1:
                apple[r][c] = 0
            else:
                tr, tc = tail.pop(0)
                snake[tr][tc] = 0
        else:
            return sol
check(cur_dir)
print(sol)


```
