```python
n,k = map(int,input().split())
word = [set(list(input())) for _ in range(n)]
```

```python
from itertools import combinations
necc = set('acint')
comb = ['b','d','e','f','g','h','j','k','l','m','o',
        'p','q','r','s','u','v','w','x','y','z']
sol = 0
if k>5:
    li = list(combinations(comb,k-5))
    for i in li:
        cnt = 0
        temp = necc
        temp = temp | set(i)
        for w in word:
            if w <= temp:
                cnt += 1
        sol = max(sol,cnt)
elif k == 5:
    cnt = 0
    for w in word:
        if w <= necc:
            cnt += 1
    sol = cnt
print(sol)
```
