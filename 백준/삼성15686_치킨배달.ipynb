```python
n, m = map(int,input().split())
city = list(list(map(int, input().split())) for _ in range(n))
```

```python
print(city)
```

```python
from itertools import combinations
chik = []
for i in range(n):
  for j in range(n):
    if city[i][j] == 2:
      chik.append([i,j])
result = list(combinations(chik, m))

print(result)

min_distance = 10000000
for i in range(len(result)):
  distance = 0
  for r in range(n):
    for c in range(n):
      if city[r][c] == 1:
        temp = 10000000
        for j in range(m):
          temp = min(temp, abs(r - result[i][j][0]) + abs(c - result[i][j][1]))
        distance += temp
  min_distance = min(min_distance, distance)
print(min_distance)
#list(combinations(items, 2))
```
