```
from collections import deque
n, m, h = map(int,input().split())
t = [[list(input().split()) for _ in range(m)] for _ in range(h)]
```

```
dx = [1, -1, 0, 0, 0, 0]
dy = [0, 0, 1, -1, 0, 0]
dz = [0, 0, 0, 0, 1, -1]
queue = deque()
for i in range(m):
  for j in range(n):
    for k in range(h):
      if t[k][i][j] == '1':
        queue.append([i, j, k])
def dfs(m, n, h, te, queue):
  s = [[[0]*n for _ in range(m)] for _ in range(h)]
  u = 0
  for i in range(m):
    for j in range(n):
      for k in range(h):
        if t[k][i][j] == '0':
          u += 1
  if u == 0:
    return(0)
  while queue:
    a, b, c = queue.popleft()
    for k in range(6):
      x = a + dx[k]
      y = b + dy[k]
      z = c + dz[k]
      if 0 <= x < m and 0 <= y < n and 0 <= z < h and t[z][x][y] == '0' and s[z][x][y] == 0:
        queue.append([x, y, z])
        t[z][x][y] = '1'
        s[z][x][y] = s[c][a][b] + 1
        s1, s2, s3 = x, y, z
  for i in range(m):
    for j in range(n):
      for k in range(h):
        if t[k][i][j] == '0':
          return(-1)
  return(s[s3][s1][s2])
print(dfs(m, n, h, t, queue))
```
