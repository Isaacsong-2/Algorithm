```
n = int(input())
t = [list(map(int, input().split())) for _ in range(n)]
```

```
if n==1:
  print(t[0][0])
else:
  k = 2
  for i in range(1,n):
    for j in range(k):
      if j == 0:
        t[i][j] = t[i][j] + t[i-1][j]
      elif i == j:
        t[i][j] = t[i][j] + t[i-1][j-1]
      else:
        t[i][j] = max(t[i - 1][j - 1], t[i - 1][j]) + t[i][j]
    k +=1
print(max(t[n-1]))
```
