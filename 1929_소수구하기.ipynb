```
import math
M, N = map(int,input().split())
N += 1
def isprime(m, n):
  prime = [True]*n
  for i in range(2, int(math.sqrt(n))+1):
    if prime[i]:
      for j in range(2*i, n, i):
        prime[j] = False

  for i in range(m, n):
    if i > 1 and prime[i] == True:
      print(i)
isprime(M, N)
```
