```
n = int(input())
p = []
for _ in range(n):
  p.append(int(input()))
```

```
if n==1:
  s = p[0]
elif n==2:
  s = [p[0], p[0]+p[1]]
else:
  s = [p[0], p[0]+p[1], max(p[0]+p[2],p[1]+p[2])]
if n == 1:
  print(s)
elif n == 2 or n == 3:
  print(s[n-1])
else:
  for i in range(3, n):
    s.append(max(s[i-2]+p[i], s[i-3]+p[i-1]+p[i]))
  print(s[n-1])
```

```
n = int(input())
p = []

for _ in range(n):
    p.append(int(input()))

def sol(n):
    if n >= 3:
        return(max(sol(n-2) + p[n], sol(n-3) + p[n-1] + p[n]))
    else:
        return(p[n])
print(sol(n-1))
```
