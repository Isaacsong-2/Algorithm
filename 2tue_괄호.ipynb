```
T = int(input())
for i in range(T):
  a = list(input()) #list로 받는다는게 무슨뜻인지..
  cnt = 0
  if (len(a) != 0):
    for j in a:
      if (j=='('):
        cnt += 1
      else:
        cnt -=1
      if cnt < 0:
        print('NO')
        break
  if (cnt==0):
    print('YES')
  if (cnt > 0):
    print('NO')     
```
